```python
import tables
import pandas as pd
import numpy as np
taste_profile = pd.read_csv('/content/drive/MyDrive/CMPE256/Project/train_triplets.txt',sep='\t',header=None,names = ['user_id','song_id','play_count'])
```

```python
taste_profile.head()
```

```python
song_metadata_file = tables.open_file("/content/drive/MyDrive/CMPE256/Project/msd_summary_file.h5", mode="r")
metadata_table = song_metadata_file.root.metadata.songs

# Link the taste profile song IDs with the metadata table
song_ids = taste_profile["song_id"].astype(str)
```

```python
metadata_table
```

```python
# Convert the table to a Pandas DataFrame
metadata_table_df = pd.DataFrame.from_records(metadata_table[:])
metadata_table_df.head()
```

```python
metadata_indices = np.searchsorted(metadata_table_df["song_id"].astype(str), song_ids)
```

```python
song_metadata = metadata_table_df.iloc[metadata_indices]
```
